```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df=pd.read_csv("delhiaqi.csv")
print(df)

print("print head")
df.head()




```

```python
# to find null value
print(df.isnull().sum())
```

```python
# to find dublicates
print(df.duplicated().sum())


```

```python
# to find error
df['date'] = pd.to_datetime(df['date'],errors='coerce')
#check is nat is exist
pd.to_datetime(..., errors='coerce')


```

```python
#remove null value
df = df.dropna(subset=['date'])
```

```python
df['month'] = df['date'].dt.month
df['day'] = df['date'].dt.day
```

```python
df['date'].head()
```

```python
#all column consist numeric value
pollutants=['co','no','no2','so2','o3','pm2_5','pm10','nh3']
df[pollutants].apply(pd.to_numeric, errors='coerce')
```

```python
# creating  a simple AQI score using PM2.5 and PM10:

df['AQI_proxy'] = (df['pm2_5']+ df['pm10'])/2
print(df)
```

 * A simplified AQI proxy was derived using the avarage of PM2.5 and PM10 concentration to analyze air quality trends.

```python
# Basic statistical analysis 

df['AQI_proxy'].describe()


```

* Mean AQI : 389.622389
* MAX AQI : 1404.735000
* MIN AQI : 64.590000

The graph shows the monthly avaragev AQI in delhi  Since the dataset contains data for only one month, the visualization display a single data point. this indicates that Aqi analysis across multiple months is not possible with the current dataset..

```python
daily_aqi = df.groupby('day')['AQI_proxy'].mean()
plt.figure(figsize=(8,5))
daily_aqi.plot(kind='line',marker='o')
plt.xlabel("Days")
plt.ylabel("Average aqi")
plt.title=('Daily Average AQI Trend in Delhi')
plt.grid()
plt.show()
```

1) Which pollutant contributes the most to Delhi’s daily AQI levels?

```python
pollutants_avg = df[pollutants].mean()
plt.figure(figsize=(8,5))
pollutants_avg.plot(kind='bar')
plt.title=('Average Pollutant Levals in Delhi')
plt.xlabel('pollutants')
plt.ylabel('Avarage contributon')
plt.grid(axis='y')
plt.show()

```

2)  How does AQI vary on a day- to- day basis during the selected month?

```python
plt.figure(figsize=(10,8))
plt.plot(daily_aqi.index,daily_aqi.values, marker='o')
plt.title=('Day-to-day Variation of Aqi in the Selected Month in delhi')
plt.xlabel('Date')
plt.ylabel('Avarage AQI')
plt.grid(True)
plt.show()
```

--> The line graph illustrates the day-to-day variation of AQI during the selected month in delhi. Fluctuations in the graph indicate changes in the air quality caused by daily factors such as traffic emmissions, whether conditions, and pollutant levels.peaks represent days with poor air quality, while lower values indicate relatively cleaner air.

3) Which days experience worst air quality in Delhi?

```python
#Daily Avrage AQI
Daily_aqi = df.groupby('day')['AQI_proxy'].mean().reset_index()

#  Sort to get worst days
Worst_days= Daily_aqi.sort_values(by='AQI_proxy',ascending=False)

#Display top worst days
Worst_days.head()
```

```python
top_worst = Worst_days.head(5)
plt.figure(figsize=(10,6))
plt.bar(top_worst['day'].astype(str),top_worst['AQI_proxy'])
plt.title=('Days with Worst Air Quality in delhi(Selected month)')
plt.xlabel('Date')
plt.ylabel('Avrage AQI')
plt.show()
```

--> The bar chart hightlights the days that experienced the worst air quality in delhi during the selected month. These days show the highest average AQI values ,indicating severe pollution levels. such peaks may be influenced by increased traffic emission, weather condition , or accumulation of pollutants.

4) How do pollutants like PM2.5 and PM10 correlate with each other? 

```python
plt.figure(figsize=(8,6))
plt.scatter(df['pm2_5'],df['pm10'])
plt.title=('Correlation between pm2.5 and pm10')
plt.xlabel('PM2.5')
plt.ylabel('PM10')
plt.grid(0)
plt.show()
```

-> The scatter plot shows a strong positive correlation between PM2.5 And PM10 levals in delhi. this indicates that when PM2.5 concentration increases, PM10 concentration also rises. the relationship suggestes common pollution sources such as vehicular emissions. construction dust, and industrial activities.

5) Are there specific daily patterns or trends in AQI during the month?

```python
plt.figure(figsize=(8,6))
plt.plot(daily_aqi.index, daily_aqi.values, marker='o')
plt.xlabel('Data')
plt.ylabel('Avrage AQI')
plt.title=('Daily AQI trend During the selected Month in delhi')
plt.grid(True)
plt.show()
```

-> The line graph reveals clear daily patterns in AQI during the selected month. AQI Levels fluctuate significantly from day to day, with noticeable spikes on certain days indicating poor air quality. These variations suggest the influence of daily factors such as traffic emmissions, weather conditions, and pollutant accumulation.

* The analysis indicates that Delhi's AQI varies significantly on a daily basis, with PM2.5 and PM10 being the major contributors to poor air quality. some days show extreme pollution levels, underscoring the need for strict air quality managment..
